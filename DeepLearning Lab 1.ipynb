```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data',header=None)
```

```python
df.shape
```

```python
df
```

```python
df.describe()
```

```python
a=[]
for i in range(0,569):
  for j in range(2,32):
    if(df[j][i]>1 or df[j][i]<0):
      a.append(j)
```

```python
res=[]
[res.append(x) for x in a if x not in res]
df[res]
```

```python
# normalization

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_values = scaler.fit_transform(df[res])
df[res] = scaled_values
df
```

```python
df.isna().any()
```

```python
x = df.iloc[:,2:]
y = df.iloc[:,1]
```

```python
x.shape
```

```python
print(x.columns)
```

```python
y = np.where(y=='M',1,0)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential,layers
```

Shallow Neural Network

```python
model =keras.Sequential([
    layers.Dense(units=16,activation='relu',input_shape=[x.shape[1]]),
    layers.Dense(units=8,activation='relu'),
    layers.Dense(units=1,activation='sigmoid') ])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(x_train,y_train,validation_data=(x_test,y_test),batch_size=10,epochs=50,verbose=1)
```

```python
history_df = pd.DataFrame(history.history)
history_df.loc[:, ['loss', 'val_loss']].plot();
print("Minimum validation loss: {}".format(history_df['val_loss'].min()))
```

```python
# Hyper Parameter Tuning
# changing units in dense layers
model =keras.Sequential([
    layers.Dense(units=8,activation='relu',input_shape=[x.shape[1]]),
    layers.Dense(units=16,activation='relu'),
    layers.Dense(units=1,activation='sigmoid') ])
```

```python
from keras.optimizers import Adam
model.compile(optimizer=Adam(lr=0.01),loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(x_train,y_train,validation_data=(x_test,y_test),batch_size=10,epochs=25,verbose=1)
# changing epochs to 25
history_df = pd.DataFrame(history.history)
history_df.loc[:, ['loss', 'val_loss']].plot();
print("Minimum validation loss: {}".format(history_df['val_loss'].min()))
```

Deep Neural Network


```python
model =keras.Sequential([
    layers.Dense(units=16,activation='relu',input_shape=[x.shape[1]]),
    layers.Dense(units=8,activation='relu'),
    layers.Dense(units=8,activation='relu'),
    layers.Dense(units=8,activation='relu'),
    layers.Dense(units=8,activation='relu'),
    layers.Dense(units=1,activation='sigmoid') ])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(x_train,y_train,validation_data=(x_test,y_test),batch_size=10,epochs=50,verbose=1)
```

```python
history_df = pd.DataFrame(history.history)
history_df.loc[:, ['loss', 'val_loss']].plot();
print("Minimum validation loss: {}".format(history_df['val_loss'].min()))
```

```python
# Hyper Parameter tuning for Deep Neural Network
model =keras.Sequential([
    layers.Dense(units=16,activation='relu',input_shape=[x.shape[1]]),
    layers.Dense(units=8,activation='relu'),
    layers.Dense(units=8,activation='relu'),
    layers.Dense(units=8,activation='relu'),
    layers.Dense(units=8,activation='relu'),
    layers.Dense(units=1,activation='sigmoid') ])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(x_train,y_train,validation_data=(x_test,y_test),batch_size=10,epochs=25,verbose=1)
```

```python
history_df = pd.DataFrame(history.history)
history_df.loc[:, ['loss', 'val_loss']].plot();
print("Minimum validation loss: {}".format(history_df['val_loss'].min()))
```
